```python
## function  damage columns to floats and creating a total damage column
def convert_damage_to_float(damage_series):
    float_series = []
    for damage in damage_series:
        if (isinstance(damage, float) | isinstance(damage, int)):
            damage_float = 0
        else:
            damage_float = float(damage[:-1])*1000
        float_series.append(damage_float)
    return float_series
storm_df['DAMAGE_PROPERTY_F'] = convert_damage_to_float(storm_df['DAMAGE_PROPERTY'])
storm_df['DAMAGE_CROPS_F'] = convert_damage_to_float(storm_df['DAMAGE_CROPS'])
storm_df['DAMAGE_TOTAL_F'] = storm_df['DAMAGE_PROPERTY_F'] + storm_df['DAMAGE_CROPS_F']
storm_df
```

```python
## these functions are to create dataframes and plots of simple bar charts
## the bar charts will also have horizontal lines for the 50th 75th and 90th percentile
def create_plt_barchart(summary_df, number_records, drop_top=0):
    col_1 = summary_df.columns[0]
    col_2 = summary_df.columns[1]
    quantile_50 = summary_df[col_2].quantile(.5)
    quantile_75 = summary_df[col_2].quantile(.75)
    quantile_90 = summary_df[col_2].quantile(.9)
    summary_df.sort_values(col_2, ascending=False, inplace=True)
    summary_df.drop(summary_df.index[:drop_top], inplace=True)
    summary_plot_df = summary_df.head(number_records)
    tick_loc = range(len(summary_plot_df[col_1]))
    plt.figure(figsize=(10,5))
    plt.bar(tick_loc, summary_plot_df[col_2],
            width=0.2,align='center')
    plt.xticks(tick_loc, summary_plot_df[col_1], rotation=90)
    plt.title(F'Top {number_records} {col_2} by {col_1}')
    plt.xlabel(col_1)
    plt.ylabel(col_2)
    plt.axhline(y=0, color='grey')
    plt.axhline(y=quantile_50, color='yellow', linewidth=0.5)
    plt.axhline(y=quantile_75, color='orange', linewidth=0.5)
    plt.axhline(y=quantile_90, color='red', linewidth=0.5)
    plt.show()

def groupby_barchart(df, groupby_column_name, agg_column_name, operation):
    df_groupby = df.groupby([groupby_column_name])
    if operation == 'sum':
        summary_df = df_groupby.sum()[[agg_column_name]]
    elif operation == 'mean':
        summary_df = df_groupby.mean()[[agg_column_name]]
    summary_df.reset_index(inplace=True)
    summary_df.columns = [groupby_column_name, f'{operation.upper()}_{agg_column_name}']
    return summary_df
    
```

```python
event_damage_sum_df = groupby_barchart(storm_df, 'EPISODE_ID', 'DAMAGE_TOTAL_F', 'sum')
create_plt_barchart(event_damage_sum_df, 20)
```

```python
event_damage_mean_df = groupby_barchart(storm_df, 'EPISODE_ID', 'DAMAGE_TOTAL_F', 'mean')
create_plt_barchart(event_damage_mean_df, 20)
```

```python
## these functions are to create dataframes and plots of simple bar charts
## the bar charts will also have horizontal lines for the 50th 75th and 90th percentile
def create_plt_scatter(summary_df):##, number_records):
    col_1 = summary_df.columns[0]
    col_2 = summary_df.columns[1]
    col_3 = summary_df.columns[2]
    col_4 = summary_df.columns[3]
    summary_plot_df = summary_df.sort_values(col_3, ascending=False)
    print(summary_plot_df)
    ##summary_plot_df = summary_df.head(number_records)
    plt.figure(figsize=(12,6))
    plt.scatter(summary_plot_df[col_2], summary_plot_df[col_3],
                    s=summary_plot_df[col_4] /2, c='maroon',
                    alpha=0.5, edgecolors='black', linewidth=1.5)
    plt.title(F'{col_3} by {col_2}')
    plt.xlabel(col_2)
    plt.ylabel(col_3)
    plt.axhline(y=0, color='grey')
    legend_data_1, = plt.plot([],[], 'or', markersize = 8, c='#778899', alpha=.5,
                              markeredgecolor = 'black', markeredgewidth =1.5)
    legend_data_2, = plt.plot([],[], 'or', markersize = 20, c='#778899', alpha=.5,
                              markeredgecolor = 'black', markeredgewidth =1.5)
    plt.legend([legend_data_1, legend_data_2],
               [f'Few {col_4}', f'Many {col_4}'],
               ncol=1, frameon=True, labelspacing = 1.2, borderpad = 1.2,
               handlelength=2, facecolor = 'white', edgecolor='#181818')
    plt.show()


    

def groupby_scatter(df, groupby_column_name, agg_column_names, operations):
    df_groupby = df.groupby([groupby_column_name])
    summary_df = df_groupby.agg({agg_column_names[0]: operations[0], agg_column_names[1]: operations[1]})
    summary_df.reset_index(inplace=True)
    ##summary_df.columns = [groupby_column_name, f'{operations[0].upper()}_{agg_column_names[0]}',
      ##                    f'{operations[1].upper()}_{agg_column_names[1]}']
    return summary_df
    
```

```python
groupby_df = groupby_scatter(storm_df, 'EVENT_TYPE', ['DAMAGE_TOTAL_F', 'EVENT_ID'], [['mean', 'sum'], ['count']])
create_plt_scatter(groupby_df)
```

```python
groupby_df = groupby_scatter(storm_df, 'EPISODE_ID', ['DAMAGE_TOTAL_F', 'EVENT_ID'], [['mean', 'sum'], ['count']])
create_plt_scatter(groupby_df)
```
